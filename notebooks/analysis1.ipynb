```python
dfcleaned2 = dfcleaned[dfcleaned['HeartProblem'] != 0]
dfcleaned2.reset_index(drop=True)
```

#### Finding the number of induviduals based on the answers for different catorires that have or haven't had Heart Problems.

```python
#sns.countplot(data= dfcleaned, x=dfcleaned.columns, y= "HeartProblem"),set(title ="Highest number of heart problems")
Diabetes1 = (dfcleaned2['Diabetes'].value_counts()[1.0])
Diabetes2 = (dfcleaned2['Diabetes'].value_counts()[2.0])
Diabetes4 = (dfcleaned2['Diabetes'].value_counts()[0.0])
Diabetes3 = Diabetes1 + Diabetes2
DiabetesFinal = pd.DataFrame({0.0:[Diabetes4],1.0:[Diabetes3]}).sort_index(ascending=True)
DiabetesFinal.set_index([['Diabetes']*len(DiabetesFinal)], inplace=True)
Diabetes = DiabetesFinal.transpose().sort_index(ascending=False)
HighBP = pd.DataFrame(dfcleaned2['HighBP'].value_counts())
HighChol = pd.DataFrame(dfcleaned2['HighChol'].value_counts())
CholCheck = pd.DataFrame(dfcleaned2['CholCheck'].value_counts())
Smoker = pd.DataFrame(dfcleaned2['Smoker'].value_counts())
Stroke = pd.DataFrame(dfcleaned2['Stroke'].value_counts())
PhysActivity = pd.DataFrame(dfcleaned2['PhysActivity'].value_counts())
HvyAlcoholConsump = pd.DataFrame(dfcleaned2['HvyAlcoholConsump'].value_counts())
AnyHealthcare = pd.DataFrame(dfcleaned2['AnyHealthcare'].value_counts())
NoDocbcCost = pd.DataFrame(dfcleaned2['NoDocbcCost'].value_counts())

newdf = pd.concat([HighBP, HighChol, CholCheck, Smoker, Stroke, Diabetes, PhysActivity, HvyAlcoholConsump, AnyHealthcare, NoDocbcCost], axis=1)
newdf
```

```python
#sns.countplot(data= newdf, x=newdf['HighBP':'NoDocbcCost'])

plotTruedata = pd.DataFrame(
    {"True": [newdf['HighBP'].iloc[0], newdf['HighChol'].iloc[0], newdf['CholCheck'].iloc[0], 
                                 newdf['Smoker'].iloc[0], newdf['Stroke'].iloc[0], newdf['Diabetes'].iloc[0], 
                                 newdf['PhysActivity'].iloc[0], newdf['HvyAlcoholConsump'].iloc[0], newdf['AnyHealthcare'].iloc[0], 
                                 newdf['NoDocbcCost'].iloc[0]]}, 
                                        index= ["HighBP", "HighChol", "CholCheck", "Smoker", "Stroke", "Diabetes", "PhysActivity", 
                                                "HvyAlcoholConsump", "AnyHealthcare", "NoDocbcCost"],
)
plotTruedata.plot(kind="bar").set(title = "People who have/had Heart Problems answered with truth/yes to health factor questions.",
                                   xlabel= "Considered health factors answered in truth/yes",
                                   ylabel= "Number of people who have/had Heart Problems")
```

#### Observing this graph, it appears as though people that have high blood pressure, high collesterol, and who are smokers all place very high in heart desease victims. This would make the most sense as these are all contributing factors to having lower health.

```python
plotFalsedata = pd.DataFrame(
{"False": [newdf['HighBP'].iloc[1], newdf['HighChol'].iloc[1], newdf['CholCheck'].iloc[1], 
                                 newdf['Smoker'].iloc[1], newdf['Stroke'].iloc[1], newdf['Diabetes'].iloc[1], 
                                 newdf['PhysActivity'].iloc[1], newdf['HvyAlcoholConsump'].iloc[1], newdf['AnyHealthcare'].iloc[1], 
                                 newdf['NoDocbcCost'].iloc[1]]}, 
                                        index= ["HighBP", "HighChol", "CholCheck", "Smoker", "Stroke", "Diabetes", "PhysActivity", 
                                                "HvyAlcoholConsump", "AnyHealthcare", "NoDocbcCost"]
)

plotFalsedata.plot(kind="bar").set(title = "People who have/had Heart Problems but who have answered negative to health factor questions.",
                                   xlabel= "Considered health factors in negation/no",
                                   ylabel= "Number of people who have/had Heart Problems")
```

#### Observing this graph it appears as though most receipents with heart problems are not heavy drinkers and many have also never had a stroke or have diabetes. This tells me heart problems aren't highly related to the likelihood of someone having a stroke or being insulin dependent.
